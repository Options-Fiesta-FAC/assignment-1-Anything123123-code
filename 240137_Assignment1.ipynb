```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
```

```python
K=float(input("Enter the value of Strike Price: "))
Premium=float(input("Enter the Premium: "))
Spot_Price_Range=float(input("Enter the Spot Price Range: "))
```

```python
#Buying A Call Option
Price_List=np.linspace(K-Spot_Price_Range,K+Spot_Price_Range,500)
```

```python
#BUYING CALL OPTIONS
PNL_Call=np.maximum(Price_List-K,0)-Premium
plt.figure(figsize=(10,6))
plt.plot(Price_List[PNL_Call>=0],PNL_Call[PNL_Call>=0],color="green",label="Profit" )
plt.plot(Price_List[PNL_Call<0],PNL_Call[PNL_Call<0],color="Red",label="loss" )
plt.axhline(0, color='black', linestyle='--', label='Break-even')
plt.xlabel("Spot Price")
plt.ylabel("Profit-Loss Buying Call")
```

```python
#SELLING CALL OPTIONS
PNL_Call_SELL=Premium-np.maximum(Price_List-K,0)
plt.figure(figsize=(10,6))
plt.plot(Price_List[PNL_Call_SELL>=0],PNL_Call_SELL[PNL_Call_SELL>=0],color="green",label="Profit" )
plt.plot(Price_List[PNL_Call_SELL<0],PNL_Call_SELL[PNL_Call_SELL<0],color="Red",label="loss" )
plt.axhline(0, color='black', linestyle='--', label='Break-even')
plt.xlabel("Spot Price")
plt.ylabel("Profit-Loss Selling Call")
```

```python
#BUYING PUT OPTIONS
PNL_Put=np.maximum(K-Price_List,0)-Premium
plt.figure(figsize=(10,6))
plt.plot(Price_List[PNL_Put>=0],PNL_Put[PNL_Put>=0],color="green",label="Profit" )
plt.plot(Price_List[PNL_Put<0],PNL_Put[PNL_Put<0],color="Red",label="loss" )
plt.axhline(0, color='black', linestyle='--', label='Break-even')
plt.xlabel("Spot Price")
plt.ylabel("Profit-Loss Selling Call")
```

```python
#Selling PUT OPTIONS
PNL_Put_SELL=Premium-np.maximum(K-Price_List,0)
plt.figure(figsize=(10,6))
plt.plot(Price_List[PNL_Put_SELL>=0],PNL_Put_SELL[PNL_Put_SELL>=0],color="green",label="Profit" )
plt.plot(Price_List[PNL_Put_SELL<0],PNL_Put_SELL[PNL_Put_SELL<0],color="Red",label="loss" )
plt.axhline(0, color='black', linestyle='--', label='Break-even')
plt.xlabel("Spot Price")
plt.ylabel("Profit-Loss Selling Call")
```
